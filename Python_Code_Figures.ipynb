```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_svg import FigureCanvasSVG

# Load data from Excel file
excel_file_path = '/Users/dylan/LucksLab Dropbox/LucksLab/Lucks_Lab_Papers/In_Progress/95_DMB_Cell_Free_Sunscreen/Python_Code_Figures/T7UVMap_RAW.xlsx'
xls = pd.ExcelFile(excel_file_path)

# Concatenate all data from different sheets to determine global vmin and vmax
all_data = pd.concat([pd.read_excel(excel_file_path, sheet_name, index_col=0) for sheet_name in xls.sheet_names])
global_vmin, global_vmax = np.nanmin(all_data.values), np.nanmax(all_data.values)

# Determine the number of sheets and set up subplots in a 2x3 grid
num_sheets = len(xls.sheet_names)
num_rows = 2
num_cols = 3
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axs array for easier indexing
axs = axs.flatten()

# Iterate through each sheet in the Excel file
for idx, sheet_name in enumerate(xls.sheet_names):
    # Load data from the current sheet
    df = pd.read_excel(excel_file_path, sheet_name, index_col=0)

    # Create a masked array to replace NaN values with -1 (or any other value outside the colormap range)
    masked_data = np.ma.masked_invalid(df.values)
    masked_data[masked_data == -1] = np.ma.masked

    # Create heatmap plot with a different colormap (e.g., 'cividis') and normalized color values
    ax = axs[idx]  # Use the current subplot
    im = ax.imshow(masked_data, cmap='magma', vmin=global_vmin, vmax=global_vmax)

    # Set ticks and labels
    ax.set_yticks(np.arange(len(df.index)))
    ax.set_xticks(np.arange(len(df.columns)))
    ax.set_yticklabels(df.index, fontsize=16)
    ax.set_xticklabels(df.columns, fontsize=16)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=0, ha="right", rotation_mode="anchor")

    # Add colorbar
    cbar = ax.figure.colorbar(im, ax=ax)

    # Set border color to black and add black boxes for NaN values
    for i in range(len(df.index)):
        for j in range(len(df.columns)):
            if np.isnan(df.iloc[i, j]):
                rect = plt.Rectangle((j - 0.5, i - 0.5), 1, 1, fill=True, facecolor="white", linewidth=0.5,
                                     edgecolor="grey")
                ax.add_patch(rect)
            else:
                rect = plt.Rectangle((j - 0.5, i - 0.5), 1, 1, fill=False, linewidth=1.5, edgecolor="black")
                ax.add_patch(rect)
                text = ax.text(j, i, '{:.2f}'.format(df.iloc[i, j]),
                               ha="center", va="center", color="white", fontsize=12)

    # Set plot title
    ax.set_title(f"Heatmap for {sheet_name}", fontsize=20)

    # Turn off grid lines
    ax.grid(False)

# Adjust layout to prevent subplot overlap
plt.tight_layout()

# Save the figure as an SVG file
svg_file = "AMNP_Plotted_Grid_Plasma_Global_Normalized_T7_RAW.svg"
canvas = FigureCanvasSVG(fig)
canvas.print_figure(svg_file, dpi=300)

# Show the entire grid of plots
plt.show()
```
